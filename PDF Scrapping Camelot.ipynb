```python
#use camelot to extract tables 1,2 *3 from the PDF
import camelot

# Replace with the path to your downloaded PDF
file_path = 'Kenya-Consumer-Price-Indices-and-Inflation-Rates-April-2025.pdf'

# Extract all tables from the PDF
tables = camelot.read_pdf(file_path, pages='all', flavor='lattice')  # Use 'stream' if tables don't have clear lines

# Print how many tables were detected
print(f"Total tables extracted: {tables.n}")
```

```python
# Let's preview table 1 (index 0)
df1 = tables[0].df
df1.head()

```

```python
# Save as CSV
df1.to_csv('Table1_Inflation_MonthlyOverview.csv', index=False)
```
