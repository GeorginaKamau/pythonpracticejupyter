```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
df = pd.read_csv("HR.csv")
```

```python
#view random dataset sample
df.sample(5)
```

```python
#null entries
df.isna().sum()
```

```python
df.info()
```

```python
#statistical summary for numeric columns
df.describe()
```

```python
# Count duplicate rows
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")

#no duplicate rows
```

```python
# Check unique values in categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns

for col in categorical_columns:
    print(f"\nUnique values in {col}:")
    print(df[col].value_counts())
```

```python
#create 1/0 binary versions of Yes/NO columns
df["Attrition_Binary"] = df["Attrition"].map({"Yes": 1, "No": 0})
df["OverTime_Binary"] = df["OverTime"].map({"Yes": 1, "No": 0})
df["Gender_Binary"] = df["Gender"].map({"Male": 0, "Female": 1})
```

```python
#label encoding tokeepcategorical columns seperate from numerical ones
from sklearn.preprocessing import LabelEncoder

categorical_cols = ["BusinessTravel", "Department", "EducationField", "JobRole", "MaritalStatus"]
for col in categorical_cols:
    df[col + "_Encoded"] = LabelEncoder().fit_transform(df[col])
```

```python
#drop unnecessary columns 
df.drop(columns=["Over18", "EmployeeNumber"], inplace=True)
```

```python
#Outlier detection
# Boxplot to visualize outliers in numerical columns
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Age', 'DailyRate', 'DistanceFromHome', 'Education']])
plt.xticks(rotation=45)
plt.title("Outlier Detection using Boxplot")
plt.show()
```

```python
#seperte the numeric columns from the categorical ones
numeric_df = df.select_dtypes(include=["int64", "float64"])
#perform correlation analysis on the numerical columns
corr_matrix = numeric_df.corr()
# Heatmap visualization
plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```
